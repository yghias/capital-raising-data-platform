# Investor Segmentation

This notebook clusters synthetic investor records into operating segments based on signal intensity and preferred fundraising stage.

```python
import pandas as pd
from sklearn.cluster import KMeans

df = pd.read_json('../sample_data/investor_signals.json')
df[['investor', 'signal_strength', 'preferred_round_type']]
```

```python
stage_map = {'Series A': 1, 'Series B': 2, 'Growth': 3}
features = pd.DataFrame({
    'signal_strength': df['signal_strength'],
    'stage_numeric': df['preferred_round_type'].map(stage_map).fillna(0)
})
model = KMeans(n_clusters=2, random_state=42, n_init='auto')
df['segment'] = model.fit_predict(features)
df[['investor', 'segment']].sort_values('segment')
```
