# Fundraising Forecasting

This notebook validates a lightweight forecasting workflow using synthetic fundraising history to estimate how qualified targets, warm introductions, and signal momentum relate to meetings booked.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_json('../sample_data/fundraising_history.json')
df
```

```python
features = df[['qualified_targets', 'warm_intro_count', 'signal_momentum']]
target = df['meetings_booked']
model = LinearRegression().fit(features, target)
pd.Series(model.coef_, index=features.columns, name='coefficient')
```
